```python
# import the required packages
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#location of my datasets
df = pd.read_csv("/content/drive/MyDrive/Deep learning/data.csv")
df.head()
```

```python
df.tail()
```

```python
plt.scatter(df[df['label'] == 0]['x1'], df[df['label'] == 0]['x2'], marker=',')

plt.scatter(df[df['label'] == 1]['x1'], df[df['label'] == 1]['x2'], marker='+')
```

```python
x_input = df[['x1','x2']].values
y_label = df[['label']].values
```

##Build a model

```python
model = Sequential()
model.add(Dense(units=1, input_dim=2, activation='sigmoid'))
model.summary()
```

##Compile the model

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

##Fit the Model

```python
model.fit(x_input, y_label, epochs=50)
```

##Evaluate the model

```python
model.evaluate(x_input, y_label)
```

##Change the Optimizer to RMSprop

```python
model1 = Sequential()
model1.add(Dense(units=1, input_dim=2, activation='sigmoid'))
model1.summary()
```

```python
model1.compile(optimizer='RMSprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model1.fit(x_input, y_label, epochs=50)
```

```python
model1.evaluate(x_input, y_label)
```

#Conclusion

In this lab Activity, when the epoch was 1000 the accuracy was 100% but when the epoch was decreased to 50 then the accuracy was 57%.
After changing the optimizer from Adam to RMSprop with epoch 50 for each optimizer, the accuracy changed from 57% to 80%
