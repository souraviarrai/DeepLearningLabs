Question2.2) Checking the desired result.

```python
#import libraries
import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
```

```python
# load dataset
iris = load_iris()
# features
X = iris.data[:, (2,3)] # petal length, petal width
# labels
y = (iris.target == 0).astype(np.int) # Iris Setosa?
per_clf = Perceptron()
per_clf.fit(X, y)
y_pred = per_clf.predict([[2, 0.5]])
print(y_pred)
```

Ans - The output came to be 0 which means false and it is correct because the target class belongs to Setosa but the input was not present in the Setosa class(i.e 0) hence it was false and it was not Setosa.

```python
iris = load_iris()
# features
X = iris.data[:, (0,1,2,3)] # petal length, petal width
# labels
y = (iris.target == 0).astype(np.int) # Iris Setosa?
per_clf = Perceptron()
per_clf.fit(X, y)
y_pred = per_clf.predict([[5.1,3.5,1.4,0.2]])
print(y_pred)
```

Ans - When we gave the four inputs for the setosa class it came to be true because the class was Setosa and the input belonged to the Setosa hence the output was 1 which means True.
